```python
import matplotlib.pyplot as plt
import numpy as np

def laplace_distribution(x, mu, ε): 
    return (1/(2*ε))*np.exp(-abs(x-mu)/ε)

x = np.linspace(-10, 10, 1000)

for ε in [0.5, 1, 2]: 
    y = laplace_distribution(x, 0, ε)
    plt.plot(x, y, label=f'ε={ε}')  

plt.title('Laplace distribution for different ε')
plt.xlabel('x')
plt.ylabel('Probability density')
plt.legend()
plt.grid(True)
plt.show()
```
